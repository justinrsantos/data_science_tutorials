# **Resizing an Image**

Images need to be in the same dimensions for processing to be done

Resizing an image using Pillow

```python
from PIL import Image
```

```python
img = Image.open('/content/imrs.php?src=https:%2F%2Farc-anglerfish-washpost-prod-washpost.s3.amazonaws.com%2Fpublic%2FHB4AT3D3IMI6TMPTWIZ74WAR54.jpg')
img_resized = img.resize((200, 200))
img_resized.save('dog_image_resized.jpg')
```

```python
img_resized1= mpimg.imread('/content/dog_image_resized.jpg')
img_plot_res=plt.imshow(img_resized1)
plt.show()
```

# **Converting RGB to grayscale using OpenCV**


```python
import cv2
img = cv2.imread('/content/imrs.php?src=https:%2F%2Farc-anglerfish-washpost-prod-washpost.s3.amazonaws.com%2Fpublic%2FHB4AT3D3IMI6TMPTWIZ74WAR54.jpg')
```

```python
type(img)
```

```python
grayscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

```python
grayscale_image.shape
```

Note that there is no 3rd argument since it is a grayscale image

```python
from google.colab.patches import cv2_imshow
```

cv2.imshow() will display the image, although this in not allowed in Google Colab

```python
cv2_imshow(grayscale_image)
```

Notice this does not display the axis unlike in matplotlib

```python
cv2.imwrite('dog_grayscale_image.jpg', grayscale_image)
```

Why would you convert colored to grayscale?

Faster processing time since colored images have 3 matrices
